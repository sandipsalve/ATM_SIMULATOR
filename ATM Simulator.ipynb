```python
class User:
  def __init__( self, pin , balance , name):
    self.correct_pin = pin
    self.balance = balance
    self.name = name
```

```python
#ATM Simulator
def atm_simulation (user):
   print(f'''
   *****Welcome { user.name} to ATM simulator*****
         ''')

   pin = int(input("Enter your pin: "))
   if pin != user.correct_pin:
    print("Incorrect pin")
    print("Thank you for using ATM")
    return

   while True:
     print ( '''chooes the following option
             1) balance,
             2) Deposit ,
             3) widraw ,
             4)  Exit ''')

     choice = int( input ( " Enter your choice"))

     if choice == 1:
       print ( " Your account balance is: " ,user.balance)

     elif choice == 2 :
       amount = int ( input ( " Enter the amount to be deposit"))
       user.balance += amount
       print ( " Your available balance is: " ,user.balance)

     elif choice == 3 :
       widraw = int ( input ( " Enter amount to be widraw "))
       if widraw > user.balance :
         print ( " Insufficient balance")
       else :
        user.balance -= widraw
        print ( " Your available balance is: " ,user.balance)
     elif choice == 4 :
        print ( " Thank you for using ATM " ," Exit")
        break
```

```python
sandip = User( 1234, 10000, "sandip")
```

```python
atm_simulation(sandip)
```
